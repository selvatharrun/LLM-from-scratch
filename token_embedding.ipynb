```python
import gensim.downloader as api
model = api.load("word2vec-google-news-300")

print(model["computer"])
```

```python
print(model['computer'].shape)  # Print first 10 dimensions
```

```python
model.most_similar(positive = ["king","man"] ,negative = ["women"])  # Find most similar words
```

```python
model.similarity("computer", "laptop")
```

```python
model.most_similar("tower", topn=5)  # Find most similar words
```
