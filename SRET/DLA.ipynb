```python
import torchvision.models as models

# Load pretrained DenseNet121
densenet = models.densenet121(pretrained=True)

# To inspect architecture
print(densenet)

# If you want to fine-tune:
for param in densenet.parameters():
    param.requires_grad = False

# Replace the classifier for your task (e.g., 10 classes)
import torch.nn as nn
densenet.classifier = nn.Linear(densenet.classifier.in_features, 10)
```

```python
import torchvision.models as models

# Load pretrained MobileNetV2
mobilenet = models.mobilenet_v2(pretrained=True)

# To inspect architecture
print(mobilenet)

# Freeze feature extractor if needed
for param in mobilenet.parameters():
    param.requires_grad = False

# Replace classifier head
mobilenet.classifier[1] = nn.Linear(mobilenet.classifier[1].in_features, 10)
```
